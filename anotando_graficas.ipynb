```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import userdata
sns.set_style('white')
```

```python
ruta = userdata.get('ruta')

df = pd.read_csv(ruta + 'diabetes-clean.csv', index_col=0)

df.head()
```

### a)

```python
# Configurar el estilo de Seaborn
sns.set(style='whitegrid')

# Crear el gráfico con los datos de glucosa
plt.figure(figsize=(10, 6))
sns.distplot(df['glucose'], kde=False, norm_hist=False)

# Añadir título y etiquetas
plt.title('Distribución de los niveles de glucosa', fontsize=16)
plt.xlabel('Niveles de Glucosa', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

# Añadir una cuadrícula
plt.grid(visible=True, linestyle='--', alpha=0.6)

# Mostrar la gráfica
plt.show()
```

```python
sns.distplot(df['glucose'], kde=False, norm_hist=False)
```

### b)

```python
pregnancy_counts = df['pregnancies'].value_counts()
```

```python
# Configurar el estilo de Seaborn
sns.set(style='whitegrid')

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
sns.barplot(x=pregnancy_counts.index, y=pregnancy_counts.values, palette='viridis', alpha=0.8)

# Añadir título y etiquetas
plt.title('Frecuencia de Número de Embarazos', fontsize=16)
plt.xlabel('Número de embarazos', fontsize=14)
plt.ylabel('Frecuencia', fontsize=14)

# Añadir una cuadrícula
plt.grid(visible=True, linestyle='--', alpha=0.6)

# Ajustar el rango de visualización en el eje Y para claridad
plt.ylim(0, max(pregnancy_counts.values) * 1.1)

# Mostrar la gráfica
plt.show()
```

```python
sns.barplot(pregnancy_counts.index, pregnancy_counts)
```

### c)

```python
healthy = df[df['outcome'] == 0]
sick = df[df['outcome'] == 1]
```

```python
# Configurar el estilo de Seaborn
sns.set(style='whitegrid')

# Crear la figura
plt.figure(figsize=(10, 6))

# Agregar las distribuciones de glucosa para cada grupo
sns.kdeplot(healthy['glucose'], shade=True, color='blue', label='Healthy', alpha=0.6, linewidth=2)
sns.kdeplot(sick['glucose'], shade=True, color='red', label='Sick', alpha=0.6, linewidth=2)

# Añadir título y etiquetas
plt.title('Distribución de Niveles de Glucosa por Estado de Salud', fontsize=16)
plt.xlabel('Edad', fontsize=14)
plt.ylabel('Densidad', fontsize=14)

# Añadir una leyenda
plt.legend(title='Estado de Salud', fontsize=12, title_fontsize=14)

# Ajustar los márgenes
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
sns.distplot(healthy['glucose'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
```

### d)

```python
# Configurar estilo
sns.set(style='whitegrid')

# Crear la figura
plt.figure(figsize=(10, 6))

# Graficar distribuciones de las variables para el grupo 'sick'
sns.distplot(sick['blood_pressure'], label='Blood Pressure', hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], label='Glucose', hist=False, kde_kws={'shade': True})
sns.distplot(sick['insulin'], label='Insulin', hist=False, kde_kws={'shade': True})

# Añadir título y etiquetas
plt.title('Distribuciones de Variables para Pacientes con Diabetes', fontsize=16)
plt.xlabel('Valor de los parametros', fontsize=14)
plt.ylabel('Densidad', fontsize=14)

# Añadir una leyenda
plt.legend(title='Variables', fontsize=12, title_fontsize=14)

# Ajustar los márgenes
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
sns.distplot(sick['blood_pressure'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['glucose'], hist=False, kde_kws={'shade': True})
sns.distplot(sick['insulin'], hist=False, kde_kws={'shade': True})
```
